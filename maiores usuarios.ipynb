```python
from os import walk
```

```python
users = []

for (dirpath, dirnames, filenames) in walk("data/users/"):
    users = dirnames
    break
```

```python
us = {}
```

```python
for user in users:
    for (dirpath, dirnames, filenames) in walk("data/users/" + user):
        us[user] = len(filenames)
        break
```

```python
uss = sorted(us.iteritems(), key=lambda (k,v): (v,k), reverse=True)
```

```python
uss
```

```python
total = 0
i = 1
for u in uss:
    print i, ": ", u
    total += u[1]
    i += 1

print total
print total / len(uss)

media = total / len(uss)
```

```python
acima = 0
abaixo = 0
igual = 0

for u in uss:
    if u[1] > media:
        acima += 1
    elif u[1] < media:
        abaixo += 1
    else:
        igual += 1

print media
print igual
print acima
print abaixo
```
